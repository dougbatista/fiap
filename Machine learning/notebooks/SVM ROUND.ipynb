```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from scipy.stats import spearmanr
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
train_data = pd.read_csv('../datasets/titanic_challenge/train.csv')
test_data = pd.read_csv('../datasets/titanic_challenge/test.csv')
```

```python
corr = train_data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
lb = LabelEncoder()

train_data_manipulation = train_data.copy()
test_data_manipulation = test_data.copy()

train_data_manipulation['Sex'] = lb.fit_transform(train_data_manipulation['Sex'])
test_data_manipulation['Sex'] = lb.fit_transform(test_data_manipulation['Sex'])
```

```python
train_data_manipulation.head()
```

```python
train_data.head()
```

```python
corr = train_data_manipulation.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
y = train_data_manipulation['Survived']

features = ['Sex', 'Parch', 'Fare', 'SibSp']

X = train_data_manipulation[features]
X_validation = test_data_manipulation[features]
```

```python
X_validation.info()
```

```python
X_validation.fillna(X_validation.median(), inplace=True)
```

```python
X_validation.info()
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=38)
```

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.fit_transform(X_test)
X_validatiion = sc.fit_transform(X_validation)

X_scaled = sc.fit_transform(X)
```

```python
X_train_scaled
```

```python
from sklearn import svm
```

```python
for i in range (1, 100):
  for j in range(1, 100):
    
    c = i
    gamma = j / 100
    
    model = svm.SVC(C=c, gamma=gamma, random_state=42)
    
    model.fit(X_train_scaled, y_train)
    
    train_preds = model.predict(X_train_scaled)
    test_preds = model.predict(X_test_scaled)
    
    SVM_dic_metrics = {'train_accuracy_score': np.round(accuracy_score(train_preds, y_train), decimals=3) * 100,
                   'test_accuracy_score': np.round(accuracy_score(test_preds, y_test), decimals=3) * 100}

    if(SVM_dic_metrics['train_accuracy_score'] >= 82 and SVM_dic_metrics['test_accuracy_score'] >= 82):
      print("C = {0}, gamma = {1}, train_accuracy = {2}, test_accuracy = {3} ".format(i, gamma, SVM_dic_metrics['train_accuracy_score'], SVM_dic_metrics['test_accuracy_score']))

    
```

```python
# C = 5, gamma = 0.44, train_accuracy = 82.0, test_accuracy = 83.2 

model = svm.SVC(C=5, gamma=0.44)
model.fit(X_scaled, y)
```

```python
predictions = model.predict(X_validatiion)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
```
