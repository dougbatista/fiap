```python
"""
Exemplo do livro:
ANTON, Howard; RORRES, Chris. Álgebra Linear com Aplicações. 10ª Edição. Porto Alegre: Bookman, 2012.

página: 296
"""
#teste dos autovalores e autovetores
import numpy as np
from numpy import array

A = array([
[3, 0],
[8, -1]])

#Autovetor [1, 2] para o autovalor 3
x = array([1, 2])
B = A.dot(x)
norma1 = np.linalg.norm(x)
norma2 = np.linalg.norm(x*3)
print (B)
print("Comprimento do vetor [1,2]: ", norma1)
print("Comprimento do vetor [3,6]: ", norma2)
print()

#Autovetor [5, 10] para o autovalor 3
x = array([5, 10])
B = A.dot(x)
norma1 = np.linalg.norm(x)
norma2 = np.linalg.norm(x*3)
print (B)
print("Comprimento do vetor [5, 10]: ", norma1)
print("Comprimento do vetor [15, 30]: ", norma2)
print()

#Autovetor [0.4472136, 0.89442719] para o autovalor 3
x = array([0.4472136, 0.89442719])
B = A.dot(x)
norma1 = np.linalg.norm(x)
norma2 = np.linalg.norm(x*3)
print (B)
print("Comprimento do vetor [0.4472136, 0.89442719]: ", norma1)
print("Comprimento do vetor [1.3416408, 2,68328157]: ", norma2)
print()

#Autovetor [0, 5] para o autovalor -1
x = array([0, 5])
B = A.dot(x)
norma1 = np.linalg.norm(x)
norma2 = np.linalg.norm(-x)
print (B)
print("Comprimento do vetor [0,5]: ", norma1)
print("Comprimento do vetor [0,-5]: ", norma2)

```
