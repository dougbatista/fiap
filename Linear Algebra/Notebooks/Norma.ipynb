```python
"""
Norma de um vetor
"""
import numpy as np

vecA = np.array((3, 4))
vecB = np.array((7, 3, 4))
vecC = np.array((21, 9, 12))
vecD = np.array((3, -3, -2, 0, -3, 13, 5))

normaA = np.linalg.norm(vecA)
normaB = np.linalg.norm(vecB)
normaC = np.linalg.norm(vecC)
normaD = np.linalg.norm(vecD)

print(vecA)
print("Norma: " , normaA)
print()
print(vecB)
print("Norma: " , normaB)
print()
print(vecC)
print("Norma: " , normaC)
print()
print(vecD)
print("Norma: " , normaD)
```

```python
"""
Cálculo do ângulo

"""
import numpy as np
import math

v = np.array((3, 4, 0))
u = np.array((1.5, -2, 2))

normaA = np.linalg.norm(u)
normaB = np.linalg.norm(v)


print(u)
print("Norma u: " , normaA)
print()
print(v)
print("Norma: " , normaB)
print()

uv = np.dot(u, v)

angulorad = math.degrees(np.arccos(np.linalg.norm(uv) / (normaA * normaB)))

print("Norma uv: ", uv)
print("Ângulo: ", angulorad, " graus")
```
