```python
# invert matrix
from numpy import array
from numpy.linalg import inv
# define matrix
A = array([
[0.0, -5.0, 4.0],
[0.0, 3.0, 6.0],
[1.0, 5.0, 1.0]])
print(A)
print()
# invert matrix
B = inv(A)
print(B)
print()
# multiply A and B
# alternative syntax in Python 3.5: I = A @ B
I = A.dot(B)
print(I)
```
